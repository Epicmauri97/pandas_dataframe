```python
import pandas as pd
import numpy as np

# Create a DataFrame with 10 rows, 4 columns, and random values
df = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])

# Introduce NaN values in some cells
df.iloc[2, 1] = np.nan
df.iloc[5, 3] = np.nan
df.iloc[7, 2] = np.nan

# Highlight NaN values
highlighted_df = df.style.applymap(lambda x: 'background-color: yellow' if pd.isna(x) else '', subset=pd.IndexSlice[:, :])

highlighted_df
```

```python
import pandas as pd
import numpy as np

# Crea un dataframe di esempio con valori casuali
np.random.seed(0)
df = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])

# Funzione per evidenziare le colonne specifiche
def highlight_columns(col):
    color = 'yellow' if col.name == 'A' else 'lightgreen' if col.name == 'C' else ''
    return [f'background-color: {color}' for _ in col]

# Applica la funzione di stilizzazione alle colonne usando apply
styled_df = df.style.apply(highlight_columns, axis=0)

# Mostra il dataframe stilizzato
styled_df
```
